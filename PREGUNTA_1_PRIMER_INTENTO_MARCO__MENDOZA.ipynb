```python
import itertools 

def permute(s, i, n): 
  if i==n: 
      print('->'.join(s))
  else: 
      for j in range(i,n): 
        s[i], s[j] = s[j], s[i] 
        permute(s, i+1, n) 
        s[i], s[j] = s[j], s[i]

def findsubsets(s, n): 
    a = list(map(lambda a: list(a), itertools.combinations(s, n)))
    return a

def getPermutations(array, n):
  ss = findsubsets(array, n)
  print(ss)
  for i in ss:
    permute(i, 0, len(i))



a = ["Gaseosa", "Cerveza", "Vino", "Agua"]

for i in range(len(a)+1):
  getPermutations(a, i)
```
